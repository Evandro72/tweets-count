```python
import pandas as pd
from os import walk
import pandas as pd
import json
import matplotlib.pyplot as plt
import re
import string
import nltk
import datetime
import seaborn as sns
```

```python
data = pd.read_csv("./Tweets.csv")
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

```python
data["hour"] = data["Date"].apply(lambda x: x.hour)
```

```python
hourly_data = data.groupby("hour").count().sort_values("hour").reset_index()
hourly_data = hourly_data.rename({"hour":"Hour","User":"Number of Tweets"},axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplots(figsize=(20,7))
sns.barplot(x = "Hour", y = "Number of Tweets", palette = 'hls' ,data = hourly_data, dodge = False)
plt.xlabel("Hour",size=15)
plt.ylabel("Number of Tweets",size=15)
plt.title("Number of tweets in each hour",size=16)

plt.show()
```
